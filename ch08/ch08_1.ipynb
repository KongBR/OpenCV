```python
img = cv2.imread('/content/gdrive/My Drive/Images/char.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations=1)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - gImg
cv2_imshow(edgeImg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/Images/milkdrop.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations=2)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - gImg
cv2_imshow(edgeImg)
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype=np.ubyte)

threshold = 145

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    if(gImg[h,w] > threshold):
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(gBin, kernel, iterations=3)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - gBin
cv2_imshow(edgeImg)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeImg = cv2.erode(gBin, kernel, iterations=2)
cv2_imshow(erodeImg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(erodeImg, kernel, iterations=5)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - erodeImg
cv2_imshow(edgeImg)
```
