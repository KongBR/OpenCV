```python
img = cv2.imread('/content/gdrive/My Drive/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
gPlus = np.zeros((gImg.shape[0], gImg.shape[1]), dtype = np.ubyte)
```

```python
val = 20

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = gImg[h,w] + val
    gPlus[h,w] = plusVal

plt.imshow(gPlus, cmap='gray')
```

```python
val = 20

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = gImg[h,w] - val
    gPlus[h,w] = plusVal

plt.imshow(gPlus, cmap='gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/Images/aero2.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gImg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg1.shape)
#plt.imshow(gImg1, cmap='gray')
cv2_imshow(gImg1)
```

```python
img = cv2.imread('/content/gdrive/My Drive/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gImg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg2.shape)
#plt.imshow(gImg2, cmap='gray')
cv2_imshow(gImg2)
```

```python
for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    plusVal = (np.int32(gImg1[h,w]) + np.int32(gImg2[h,w]))/2
    if(plusVal >255):
      plusVal = 255
    gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```

```python
weight = 0.6

for h in range(gImg1.shape[0]):
  for w in range(gImg1.shape[1]):
    plusVal = np.int32(gImg1[h,w])*weight + np.int32(gImg2[h,w])*(1-weight)
    if(plusVal >255):
      plusVal = 255
    gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```

```python
for h in range(gImg2.shape[0]):
  for w in range(gImg2.shape[1]):
    plusVal = (np.int32(gImg2[h,w])*(1/8) + np.int32(gImg2[h,w])*(2/8) + np.int32(gImg2[h,w])*(3/8) + np.int32(gImg2[h,w])*(4/8) + np.int32(gImg2[h,w])*(5/8) + np.int32(gImg2[h,w])*(6/8) + np.int32(gImg2[h,w])*(7/8) + np.int32(gImg2[h,w])*(8/8))/8
    if(plusVal >255):
      plusVal = 255
    gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```
